```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('trainhpp.csv')
```

```python
del train['Id']
train['SalePrice'] = train['SalePrice'].astype(int)
df = train
df
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
df = df.drop((missing_data[missing_data['Total'] > 1]).index,1)
df = df.drop(df.loc[df['Electrical'].isnull()].index)
df = df.dropna()
df.shape
```

```python
from sklearn import preprocessing
def number_encode_features(init_df):
    result = init_df.copy() 
    encoders = {}
    for column in result.columns:
        if result.dtypes[column] == np.object:
            encoders[column] = preprocessing.LabelEncoder()
            result[column] = encoders[column].fit_transform(result[column])
    return result, encoders
encoded_data, encoders = number_encode_features(df) 
encoded_data.head()
```

```python
from sklearn.model_selection import train_test_split
X = encoded_data.iloc[: , encoded_data.columns!='SalePrice']
y =  encoded_data['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size= 0.3, random_state=42)
```

```python
X
```

```python
from sklearn import svm
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
r2_score={}
mae={}
mse={}
rmse={}
count=0
regressor=['Linear SVM','Muliple Linear Regression','Decision Tree Regressor','Random Forest Regressor','Ridge Regression','Lasso Regression']
models=[svm.SVR(kernel='linear'),LinearRegression(),DecisionTreeRegressor(),RandomForestRegressor(n_estimators=1000), Ridge(alpha=4),Lasso(alpha=0.0001)]
for i in models:    
    model = i
    model.fit(X_train,y_train)
    prediction=model.predict(X_test)
    temp1={regressor[count]:model.score(X_test,y_test)}
    temp2={regressor[count]:metrics.mean_absolute_error(y_test,prediction)}
    temp3={regressor[count]:metrics.mean_squared_error(y_test,prediction)}
    temp4={regressor[count]:metrics.mean_squared_error(y_test,prediction,squared=False)}
    r2_score.update(temp1)
    mae.update(temp2)
    mse.update(temp3)
    rmse.update(temp4)
    count+=1
```

```python
from sklearn.ensemble import GradientBoostingRegressor

reg = GradientBoostingRegressor(random_state=42,
                                n_estimators = 1000,
                                max_depth = 5,
                                min_samples_leaf = 6,
                                min_samples_split = 10)

reg.fit(X_train, y_train)
pred_reg = reg.predict(X_test)
temp1={'Gradient Boosting Regressor':reg.score(X_test,y_test)}
temp2={'Gradient Boosting Regressor':metrics.mean_absolute_error(y_test,pred_reg)}
temp3={'Gradient Boosting Regressor':metrics.mean_squared_error(y_test,pred_reg)}
temp4={'Gradient Boosting Regressor':metrics.mean_squared_error(y_test,pred_reg,squared=False)}
r2_score.update(temp1)
mae.update(temp2)
mse.update(temp3)
rmse.update(temp4)
```

```python
from xgboost import XGBRegressor
from sklearn.preprocessing import LabelEncoder

xgb = XGBRegressor(learning_rate=0.01,
                       n_estimators=1000,
                       max_depth=4,
                       min_child_weight=0,
                       gamma=0.6,
                       subsample=0.7,
                       colsample_bytree=0.7,
                       objective='reg:linear',
                       nthread=-1,
                       scale_pos_weight=1,
                       seed=27,
                       reg_alpha=0.00006,
                       random_state=42)

xgb.fit(X_train, y_train, verbose=False)
pred_xgb = xgb.predict(X_test)
temp1={'XGBoost Regressor':xgb.score(X_test,y_test)}
temp2={'XGBoost Regressor':metrics.mean_absolute_error(y_test,pred_xgb)}
temp3={'XGBoost Regressor':metrics.mean_squared_error(y_test,pred_xgb)}
temp4={'XGBoost Regressor':metrics.mean_squared_error(y_test,pred_xgb,squared=False)}
r2_score.update(temp1)
mae.update(temp2)
mse.update(temp3)
rmse.update(temp4)
```

```python
from catboost import CatBoostRegressor
cat_model = CatBoostRegressor(iterations = 1000,
                              random_state = 42)
cat_model.fit(X_train,y_train, verbose=False)
cat_pred = cat_model.predict(X_test)
temp1={'CATBoost Regressor':cat_model.score(X_test,y_test)}
temp2={'CATBoost Regressor':metrics.mean_absolute_error(y_test,cat_pred)}
temp3={'CATBoost Regressor':metrics.mean_squared_error(y_test,cat_pred)}
temp4={'CATBoost Regressor':metrics.mean_squared_error(y_test,cat_pred,squared=False)}
r2_score.update(temp1)
mae.update(temp2)
mse.update(temp3)
rmse.update(temp4)
```

```python
comparison = pd.DataFrame.from_dict(r2_score, orient ='index')  
comparison = comparison.rename(columns={0:'R2 Score'})
comparison
```

```python
r2_list=list(r2_score.values())
mean_absolute=list(mae.values())
mean_squared=list(mse.values())
root_mean=list(rmse.values())
comparison['Mean Absolute Error']=mean_absolute
comparison['Mean Squared Error']=mean_squared
comparison['Root MSE']=root_mean
```
